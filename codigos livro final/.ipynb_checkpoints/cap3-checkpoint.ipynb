```python
import pandas as pd
df = pd.read_csv("dados_ecommerce.csv")
df.head()
```

```python
df.info()
```

```python
import numpy as np

df['Avaliação'] = df['Avaliação'].replace('s/n', np.nan)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Preço'], inplace=True)
```

```python
df['Categoria'] = df['Categoria'].fillna("Sem Categoria")
```

```python
df['Avaliado'] = ~df['Avaliação'].isnull()
```

```python

media_cat = df.groupby('Categoria')['Avaliação'].mean()
df['Avaliação'] = df['Avaliação'].fillna(df['Categoria'].map(media_cat))

df
```

```python
df['Avaliação'] = pd.to_numeric(df['Avaliação'], errors='coerce')
```

```python
df.info()
```

```python

media_cat = df.groupby('Categoria')['Avaliação'].mean()
df['Avaliação'] = df['Avaliação'].fillna(df['Categoria'].map(media_cat))
```

```python
df.loc[df['Categoria'] == 'eletronicos', 'Categoria'] = 'Eletrônicos'
```

```python

media_cat = df.groupby('Categoria')['Avaliação'].mean()
df['Avaliação'] = df['Avaliação'].fillna(df['Categoria'].map(media_cat))
```

```python
df['Avaliação'] = df['Avaliação'].clip(lower=1, upper=5)
```

```python
df['Quantidade'] = df['Quantidade'].clip(lower=0)
```

```python
df['Produto'] = df['Produto'].str.lower().str.replace(' ', '')



df
```

```python
df.drop_duplicates(subset=['Produto'], keep='first', inplace=True)
```

```python
df
```

### Transformação

```python
from sklearn.preprocessing import MinMaxScaler

numerical_columns = df.select_dtypes(include=['int', 'float']).columns
scaler = MinMaxScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
df
```

```python
mapa = {'Alimentos': 1, 'Brinquedos': 2, 'Eletrônicos': 3, 'Vestuário': 4, 'Sem Categoria': 5}
df['Categoria'] = df['Categoria'].map(mapa)
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['Categoria'])
```

```python
df
```

```python
df['discretizado'] = pd.cut(df['Preço'], bins=3, labels=["Baixo", "Médio", "Alto"])
```

```python
df
```

```python
sampled_df = df.sample(frac=0.1)
sampled_df
```

```python
irrelevantes = ['Produto']
df.drop(irrelevantes, axis=1, inplace=True)
```

```python
from sklearn.decomposition import PCA

# Inicializando PCA para reduzir os dados a 2 componentes
pca = PCA(n_components=2)

# Supondo df como um DataFrame com muitas colunas numéricas
reduced_df = pd.DataFrame(pca.fit_transform(df[numerical_columns]), columns=['PC1', 'PC2'])
```
